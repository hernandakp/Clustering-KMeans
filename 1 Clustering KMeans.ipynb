```python
'Author: hernanda472@gmail.com'

import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_excel("1 Dataset.xlsx")
df
```

```python
plt.figure(figsize=(6, 9))
plt.title('Proporsi Individu yang Menggunakan Internet di Menurut Provinsi Tahun 2017-2019')
plt.scatter(df.Tahun_2017, df['Provinsi'], color = 'blue', label = "2017")
plt.scatter(df.Tahun_2018, df['Provinsi'], color = 'red', label = "2018")
plt.scatter(df.Tahun_2019, df['Provinsi'], color = 'green', label = "2019")
plt.xlabel('Persentase')
plt.ylabel('Provinsi')
plt.legend()
```

```python
plt.figure(figsize=(12, 8))
plot_2017 = sns.distplot(df["Tahun_2017"], color = 'blue', label = "2017")
plot_2018 = sns.distplot(df["Tahun_2018"], color = 'red', label = "2018")
plot_2019 = sns.distplot(df["Tahun_2019"], color = 'green', label = "2019")
plt.xlabel('Persentase')
plt.legend()
```

```python
tahun_2017 = df['Tahun_2017'].values
tahun_2019 = df['Tahun_2019'].values
X = np.array(list(zip(tahun_2017, tahun_2019)))
plt.figure(figsize=(5, 5))
plt.scatter(tahun_2017, tahun_2019, c = 'black')
plt.xlabel('Tahun 2017')
plt.ylabel('Tahun 2019')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward' )
y_hc = hc.fit_predict(X)
plt.figure(figsize=(5, 5))
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], c = 'purple', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], c = 'green', label = 'Cluster 3')
plt.title('Clusters')
plt.xlabel('Tahun 2017')
plt.ylabel('Tahun 2019')
plt.legend()
plt.show()
```

```python
km3=KMeans(n_clusters=3,init='k-means++', n_init=10, random_state=0)
y_means = km3.fit_predict(X)
plt.figure(figsize=(5, 5))
plt.scatter(X[y_means==0,0],X[y_means==0,1], c='blue',label='Cluster1')
plt.scatter(X[y_means==1,0],X[y_means==1,1], c='purple',label='Cluster2')
plt.scatter(X[y_means==2,0],X[y_means==2,1], c='green',label='Cluster3')
plt.scatter(km3.cluster_centers_[:,0], km3.cluster_centers_[:,1],s=200,marker='s', c='red', alpha=0.7, label='Centroids')
plt.title('Segments')
plt.xlabel('Tahun 2017')
plt.ylabel('Tahun 2019')
plt.legend()
plt.show()
```
